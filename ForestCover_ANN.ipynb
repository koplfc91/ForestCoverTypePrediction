```python
import numpy as np
import pandas as pd                # Data exploration, manipulation functions
import matplotlib.pyplot as plt
import seaborn as sns 
```

```python
pd.options.display.max_columns = 999
```

```python
forest_train = pd.read_csv("train.csv") 
```

```python
forest_train.head()
```

```python
forest_train.info()
```

```python
forest_train.shape
```

```python
covertype_train = forest_train['Cover_Type']
data_train = forest_train.drop(['Id','Cover_Type'],axis = 1)
```

```python
type(covertype_train)
```

```python
forest_test = pd.read_csv("test.csv")
data_test = forest_test.drop("Id",axis = 1)
```

```python
data_test.info()
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
sc = StandardScaler()
```

```python
data_train = sc.fit_transform(data_train)
data_test = sc.transform(data_test)
```

```python
import keras
```

```python
from keras.models import Sequential
from keras.layers import Dense
```

```python
classifier = Sequential()
```

```python
classifier.add(Dense(units = 4,activation = 'relu',kernel_initializer = 'uniform',input_dim = 54)) #First Hidden Layer
```

```python
classifier.add(Dense(units = 8, kernel_initializer = 'uniform', activation = 'sigmoid')) # Output Layer. 8 units needed - 1 for each category and 1 extra as to_categorical creates an extra 0 class. Wont impact output as 0 is not present in training set at all
```

```python
classifier.compile(optimizer = 'adam', loss = 'categorical_crossentropy', metrics = ['accuracy'])
```

```python
covertype_train
```

```python
from keras.utils.np_utils import to_categorical
```

```python
categorical_covertype_train = to_categorical(covertype_train)#`categorical_crossentropy` expects targets to be binary matrices (1s and 0s) of shape (samples, classes)
categorical_covertype_train[0:10]
```

```python
categorical_covertype_train.shape
```

```python
categorical_covertype_train[0:100]
```

```python
#new_categorical_covertype_train = np.delete(categorical_covertype_train,0,axis=1)  # Dropping the extra column that got created. to_categorical converts int array from 0 to num_classes.Have to delete first column so put 0 and not 1
## Cannot use this approach as predict is predicting class 0 which doesnt exist and excluding class 7
```

```python
#new_categorical_covertype_train[0:100]
```

```python
classifier.fit(data_train, categorical_covertype_train, batch_size = 1000, epochs = 200)
```

```python
pred_test = classifier.predict_classes(data_test)   # .predict gives probabilites and not classes
```

```python
pred_test
```

```python
forest_test['Cover_Type'] = pred_test   # The basic model with 1 hidden layer with 4 units and Training with 200 epochs gave a kaggle score of 0.53 
forest_test[['Id','Cover_Type']].to_csv("ForestCover_ANN_model.csv",index = False)
```
